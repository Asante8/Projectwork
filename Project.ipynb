```python
a=np.array(data['free sulfur dioxide'])
b=np.array(data['total sulfur dioxide'])
mp.plot(a,b)
mp.title('A graph of free sulfur dioxide against total sulfur dioxide ')
mp.xlabel('free sulfur dioxide')
mp.ylabel('total sulfur dioxide')
```

```python
data.corr()
```

```python
mean_FA = data['fixed acidity'].mean()
mean_VA = data['volatile acidity'].mean()
mean_CA = data['citric acid'].mean()
mean_RS = data['residual sugar'].mean()
mean_CH = data['chlorides'].mean()
mean_FSD = data['free sulfur dioxide'].mean()
mean_TSD = data['total sulfur dioxide'].mean()
mean_D = data['density'].mean()
mean_pH = data['pH'].mean()
mean_S = data['sulphates'].mean()
mean_A = data['alcohol'].mean()
mean_Q = data['quality'].mean()
```

```python
print(f'The mean of fixed acidity is :{mean_FA}')
print(f'The mean of volatile acidity is :{mean_VA}')
print(f'The mean of citric acid is :{mean_CA}')
print(f'The mean of residual sugar is :{mean_RS}')
print(f'The mean of chlorides is :{mean_CH}')
print(f'The mean of free sulfur dioxide is :{mean_FSD}')
print(f'The mean of total sulfur dioxide is :{mean_TSD}')
print(f'The mean of density is :{mean_D}')
print(f'The mean of pH is :{mean_pH}')
print(f'The mean of sulphates is :{mean_S}')
print(f'The mean of alcohol is :{mean_A}')
print(f'The mean of quality is :{mean_Q}')
```
